```python
class generator_z:
    def __init__(self, file, regime, batch_size):
        self.file = file
        self.regime = regime
        self.batch_size = batch_size
        with h5.File(self.file, 'r') as hf:
            self.num = hf[self.regime + '/data/data'].shape[0]
        self.batch_num = self.num // self.batch_size
        self.gen_num = self.batch_num * self.batch_size
    def __call__(self):
        seq = [j for j in range(self.batch_num)]
        with h5.File(self.file, 'r') as hf:
            for i in seq:
                data = hf[self.regime + '/data/data'][i * self.batch_size:(i + 1) * self.batch_size,:,4]
                labels = np.zeros((self.batch_size, 1))
                ids = hf[self.regime + '/ev_ids/data'][
                      i * self.batch_size:(i + 1) * self.batch_size]  # id of event - starting with 'nu' or 'mu'
                ids = np.array([i[0] for i in ids])
                labels[:,0] = np.where(ids == 110, 1, 0)  # 110 - byte code for 'n'
                yield (data, labels)
```

```python
class linear_classifier():
    def __init__(self):
        pass

    def predict(self, gen, num_of_batches, pos_coef = 0.2, neg_coef = -0.06):
        preds = np.array([]).reshape(0,1)
        labels = np.array([]).reshape(0,1)
        for i in range(num_of_batches):
            clear_output(wait=True)
            print('Max number of batches =', num_of_batches)
            print('Batch #', i)
            z_array, labels_batch = next(gen)
            for zs in z_array:
                zs = zs[np.where(zs<3.0)[0]]
                zs = zs.reshape(len(zs),1)
                
                ax = np.array([i for i in range(len(zs))]).reshape(len(zs),1)
                reg = linear_model.LinearRegression()
                reg.fit(ax,zs)
                if reg.coef_[0,0]>=pos_coef:
                    preds = np.append(preds, np.array(1).reshape(1,1), axis=0)
                elif reg.coef_[0,0]<neg_coef:
                    preds = np.append(preds, np.array(0).reshape(1,1), axis=0)
                else:
                    preds = np.append(preds, np.array(0.5).reshape(1,1), axis=0)
                
            labels = np.append(labels, labels_batch, axis =0)
        return preds, labels
    
    def predict_coefs(self, gen, num_of_batches, verbose = True):
        coefs = np.array([]).reshape(0,1)
        labels = np.array([]).reshape(0,1)
        for i in range(num_of_batches):
            if verbose:
                clear_output(wait=True)
                print('Max number of batches =', num_of_batches)
                print('Batch #', i)
            z_array, labels_batch = next(gen)
            for zs in z_array:
                zs = zs[np.where(zs<3.0)[0]]
                zs = zs.reshape(len(zs),1)
                
                ax = np.array([i for i in range(len(zs))]).reshape(len(zs),1)
                reg = linear_model.LinearRegression()
                reg.fit(ax,zs)
                coefs = np.append(coefs, np.array(reg.coef_[0,0]).reshape(1,1), axis=0)
                
            labels = np.append(labels, labels_batch, axis =0)
        return coefs, labels
```

```python
class analysis():
    def __init__(self, preds, labels):
        self.preds = preds
        self.labels = labels
        self.d = self.preds - self.labels
        self.true = np.where(self.d==0)[0]
        self.false = np.where(np.abs(self.d)==1)[0]
        
        self.fp = self.false[np.where(self.d[self.false] == 1)[0]]
        self.fn = self.false[np.where(self.d[self.false] == -1)[0]]
        self.tp = self.true[np.where(self.preds[self.true] == 1)[0]]
        self.tn = self.true[np.where(self.preds[self.true] == 0)[0]]
        
        self.not_classified = np.where(self.preds == 0.5)[0]
        
        self.acc = len(self.true)/(len(self.false)+len(self.true))
    
    def prec(self, particle = 'nu'):
        if particle == 'nu':
            return len(self.tp)/(len(self.tp)+len(self.fp))
        elif particle == 'mu':
            return len(self.tn)/(len(self.tn)+len(self.fn))
        else:
            print("Particle argument should be str 'mu' or 'nu'")
    
    def rec(self, particle = 'nu'):
        if particle == 'nu':
            return len(self.tp)/(len(self.tp)+len(self.fn))
        elif particle == 'mu':
            return len(self.tn)/(len(self.tn)+len(self.fp))
        else:
            print("Particle argument should be str 'mu' or 'nu'")
            
    def f1(self, particle = 'nu'):
        if particle == 'nu':
            recall = len(self.tp)/(len(self.tp)+len(self.fn))
            precis = len(self.tp)/(len(self.tp)+len(self.fp))
            return 2*recall*precis/(recall+precis)
        elif particle == 'mu':
            recall = len(self.tn)/(len(self.tn)+len(self.fp))
            precis = len(self.tn)/(len(self.tn)+len(self.fn))
            return 2*recall*precis/(recall+precis)
        else:
            print("Particle argument should be str 'mu' or 'nu'")
    
    def plot_z(self, file, regime, idxs, N = 10):
        fig = plt.figure(figsize=(15, 8))
        with h5.File(file, 'r') as hf:
            data = hf[regime + '/data/data'][idxs[0:N],:,4:5]
            for z in data:
                z = z[np.where(z<3.0)[0]]
                ax = np.array([i for i in range(len(z))]).reshape(len(z),1)
                plt.plot(ax,z)
        plt.xlabel('Hits', fontsize=15)
        plt.ylabel('Z - coordinate', fontsize=15)
        plt.title('Events', fontsize=15)
        plt.grid(ls = ':')
        plt.show()
```

```python
path_to_coefs_ds = './ds_coefs_'+name[:-3]+ '.h5'
with h5.File(path_to_coefs_ds, 'w') as hfc:
    for regime in ['train','test','val']:
        with h5.File(path_to_h5, 'r') as hf:
            L = len(hf[regime + '/data/data'])
            max_num_of_batches = L//batch_size
            g = generator_z(file = path_to_h5, regime = regime, batch_size = batch_size)()
            for n in range(max_num_of_batches):
                coefs, labels = LC.predict_coefs(g, num_of_batches = 1, verbose = False)
                if n == 0:
                    hfc.create_dataset(regime+'/data/coefs', data=coefs, maxshape=(None,1))
                    hfc.create_dataset(regime+'/data/labels', data=labels, maxshape=(None,1))
                else:
                    hfc[regime+'/data/coefs'].resize((hfc[regime+'/data/coefs'].shape[0] + coefs.shape[0]), axis = 0)
                    hfc[regime+'/data/coefs'][-coefs.shape[0]:] = coefs
                    
                    hfc[regime+'/data/labels'].resize((hfc[regime+'/data/labels'].shape[0] + labels.shape[0]), axis = 0)
                    hfc[regime+'/data/labels'][-labels.shape[0]:] = labels
                clear_output(wait=True)
                print('Current regime =', regime)
                print('Max num of batches =', max_num_of_batches)
                print('Batch #', n)
```

```python
def make_preds(coefs, pos_coef = 0.05, neg_coef = -0.05): #pos_coef = 0.52, neg_coef = -0.42
    idxs_nu = np.where(coefs[:]>=pos_coef)[0]
    idxs_mu = np.where(coefs[:]<neg_coef)[0]
    preds = np.ones((len(coefs),1))/2
    preds[idxs_nu] = 1.
    preds[idxs_mu] = 0.
    return preds
```
